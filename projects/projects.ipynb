#### Оптимальные представители классов гомологий как задача линейного программирования
Аннотация

Диаграмму устойчивости можно дополнить информацией о представителях каждого $k$-класса гомологий в виде цепей соответствующих $k$-симплексов. Оптимальные представители классов гомологий можно получить решая задачу оптимизации как нахождения цепи гомологичной данной, например полученной стандартный алгоритм приведения матрицы (фильтрованного) комплекса к верхнетреугольному виду, минимизирующий тот или иной функционал — оптимальных в смысле количества симплексов в цепи либо суммы  на них значений функции фильтрации.

Цели и задачи
- изучить теорию (представители классов гомологий, в том числе устойчивых, задача линейного программирования для нахождения минимальных представителей);
- найти/реализовать вычисление оптимальных циклов, применить к облакам точек ДЗ1 и/или (случайным) графам;
- сделать доклады по теории/постановке задачи, и по полученным результатам.

Список литературы
- [Li2021] Minimal Cycle Representatives in Persistent Homology Using Linear Programming: An Empirical Study With User’s Guide

#### Нейросетевые модели на множествах для обучения на диаграммах устойчивости 
Аннотация

Многие задачи машинного обучения могут быть поставлены как обучение на множествах, в том числе обучение на диаграммах устойчивости представляющих собой множества переменной мощности. Рассмотренная в ДЗ2 модель Deep Sets, в которой кодировщик применяется к каждому вектору множества может быть представлена как частный случай более общей модели Janossy Pooling в которой кодировщик применяется ко всем перестановкам элементов множества. Также, в частном случае перестановок пар элементов множества модель Janossy Pooling [Murphy2018] совпадает с другой моделью для обучения на множествах — Relational Networks [Santoro2017].

Цели и задачи
- изучить теорию (инвариантность/эквивариантность к перестановкам, универсальная аппроксимация функций для множеств и т.д.);
- найти/реализовать модель Janossy pooling в различных вариантах (k-ary, sampling);
- обучить модель на диаграммах облаков точек/графов из ДЗ2, сравнить с базовыми моделями Deep Sets/Transformer;
- сделать доклады по теории/постановке задачи, и по полученным результатам.

Список литературы
- [Wagstaff2021] Universal Approximation of Functions on Sets
- [Murphy2018] Janossy Pooling: Learning Deep Permutation-invariant Functions for Variable-size Inputs
- [Santoro2017] A Simple Neural Network Module for Relational Reasoning

#### Симплеты: мотивы в симплициальных комплексах как базовые признаки для нейросетевых моделей на комплексах
Аннотация

Графы и комплексы можно описать количеством вхождений в них элементарных подграфов (графлетов) и подкомплексов (симплетов [Malod-Dognin2019,Kim2023]), а каждую вершину - вектором принадлежности вершины определенному мотиву либо орбите. Вектора таких признаком могут быть использованы в качества базовых для графовых нейросетевых моделей и трансформеров на графах.

Цели и задачи
- изучить теорию (мотивы и орбиты в графах и комплексах а также алгоритмы их поиска, система линейных уравнений для мотивов и т.д.);
- найти/реализовать поиск симплициальных мотивов в комплексах, как переборную, так и с использованием системы линейных уравнений (ORCA [Ribeiro2019, Section 3.2.1]);
- реализовать нейросеть на комплексах, обобщение нейросети на графах и/или трансформер с использованием числа вхождения симплексов в мотив/орбиту в качестве базового признака, обучить модель на графах из ДЗ2;
- сделать доклады по теории/постановке задачи, и по полученным результатам.

Список литературы
- [Malod-Dognin2019] Functional Geometry of Protein Interactomes
- [Kim2023] Characterization of Simplicial Complexes by Counting Simplets Beyond Four Nodes
- [Ribeiro2019] A Survey on Subgraph Counting: Concepts, Algorithms, and Applications to Network Motifs and Graphlets
